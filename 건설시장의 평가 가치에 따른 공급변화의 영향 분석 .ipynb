```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_csv("전국 주거용 인허가 현황표.csv", encoding='cp949')
d1.head()
```

```python
d2 = pd.read_csv("전국 주택용 건설 수주액표.csv", encoding='cp949')
d2.head()
```

```python
d3 = pd.read_csv("전국 미분양 현황표.csv", encoding='cp949')
d3.head()
```

```python
d1.plot(x='Unnamed: 0', figsize=(15, 5), grid=True)
plt.title("건축허가현황(주거)")
plt.show()
```

```python
d2.plot(x='Unnamed: 0', figsize=(16, 5), grid=True)
plt.title("국내건설수주액(주택)")
plt.show()
```

```python
d3.plot(x='Unnamed: 0', figsize=(16, 5), grid=True)
plt.title("미분양주택현황(전국)")
plt.show()
```

```python
#d1값과 d2값의 산술연산을 위한 merge()
md1 = pd.merge(d1, d2)
md1
```

```python
###이상적인 건설시장 평가 방법
###국내건설수주액 / 건축허가현황 = 1개의 건축허가당 수주액
```

```python
md1['이상적 시장가치'] = md1['주택'] / md1['주거용']
md1
```

```python
md1['이상적 시장가치'].unique()
```

```python
#값을 소수점 둘째자리에서 반올림 ㅠ
r1 = round(md1['이상적 시장가치'], 2)
r1
```

```python
r1.plot(x='Unnamed: 0', figsize=(16, 10), grid=True)
plt.title("이상적 시장가치")
plt.show()
```

```python
### 현실적인 건설회사 평가 방법
"""
1. 2011년도~2020년도 대한민국 주거 비율
   [참조, KOSIS, 주거실태조사(2011~2020)]
   실거주 대상 조사의 한계가 있지만, 해당 비율과 아파트와 주택 등의 주거시설 각각의 건설비율이 
   동일하다는 것을 전제함.

2. 위 아파트 주거 비율을 토대로, 주택용 건설 허가수를 아파트 건설 허가수(s1)로 변환했고
   (전국 미분양 현황/s1)*100%의 계산식으로 미분양률(s2) 계산을 진행함.

3. (1-s2)*(이상적 시장평가) = 1개의 건축허가당 수주액 분석

4. 이를 통해, 국내 건설시장의 현실적 시장평가 가치를 계산해봄.

"""
###
```

```python
d4 = pd.read_csv("현실적 시장가치.csv", encoding = 'cp949')
d4.head()
```

```python
d4.plot(x='Unnamed: 0', figsize=(16, 10), grid=True)
plt.title("현실적 시장가치")
plt.show()
```

```python
md2 = pd.merge(md1, d4)
md2
```

```python
md2.drop('주택', axis = 1, inplace = True)
md2
```

```python
md2.drop('주거용', axis = 1, inplace = True)
md2
```

```python
md2.plot(x='Unnamed: 0', figsize=(16, 15), grid=True)
plt.title("이상적 시장가치와 현실적 시장가치 비교")
plt.show()
```

```python
''' 
이상적인 건설시장의 평가와 현실적인 건설시장의 평가를 비교했을때
오차율이 1%미만, 최대 약 10%이다.
오차율의 값이 커질수록 미분양율이 높다는 것을 의미하는데, 
이는 '미분양주택현황(전국)' 데이터에서 확인할 수 있다. 
'''
```
